##### <font color="blue">*Quantum Harmonic Oscillator*

***Master: Physics Foundation - Quantum Harmonic Oscillator and Quantum Operators***

> <img src="https://upload.wikimedia.org/wikipedia/commons/9/9e/HarmOsziFunktionen.png" width="400">  <img src="https://upload.wikimedia.org/wikipedia/commons/e/ea/Simple_Harmonic_Motion_Orbit.gif" width="350">

* <font color="blue">**Hamiltonian**: $\hat{H} \propto \hat{P}^2 + \hat{Q}^2$  $ = \hbar \omega \left(\hat{a}^\dagger \hat{a} + \frac{1}{2}\right) = \hbar \omega (\hat{n} + \frac{1}{2})$. $\hat{H}$ is generator of basis state rotations in phase space ($Q / P$)</font>
  * Ladder Operator Form: $\hat{H}=\hbar \omega\left(a^{\dagger} a+\frac{1}{2}\right)$. $\hat{a}$ and $\hat{a}^\dagger$ (Ladder Operators):  jump Up and Down the energy ladder.  Operators $\hat{Q}$ and $\hat{P}$ linear combinations of $\hat{a}, \hat{a}^\dagger$. Their sum of squares ($\hat{Q}^2 + \hat{P}^2$) gives Number Operator $\hat{n}$ for total energy.
  * Position/Momentum Form $\hat{H}=\frac{p^2}{2 m}+\frac{1}{2} m \omega^2 x^2$. [Quantum Harmonic Oscillation](https://en.wikipedia.org/wiki/Quantum_harmonic_oscillator) describes physical swap Kinetic $\leftrightarrow$ Potential = Time Evolution. Eg $t=0$: state in Position $Q$. Time evolution by $t=\frac{\pi}{2\omega}$ rotated $90^\circ$ to Momentum $P$ $\rightarrow$ is QFT!
  * Everything in quantum physics can be mapped onto a ['Phase Space'](https://en.wikipedia.org/wiki/Phase_space) of position $Q$ and momentum $P$ ([Formulation](https://en.wikipedia.org/wiki/Phase-space_formulation))
* <font color="blue">**From Quantum Harmonic Oscillator to Quantum Operators via $e^{-i \hat{G} \theta}$**</font>
  * $\hat{U} = e^{-i \hat{G} \theta}$ describes movement of state in terms of $Q$ and $P$. $e^{i \theta}$ represents point on unit circle, keeps it unitary ($U^{\dagger}U=I$).
  * Set $-i$ and add generator $G$ ensures transformation is "rotation" in Hilbert space that preserves length (norm) of state vector.
  * <font color="blue">$e^{-i \hat{G} \theta}$ versus $e^{-i \hat{H} t / \hbar}$</font>: $\hat{U} = e^{-i \hat{G} \theta}$ is the general form. $\hat{G}$ is a transformation. Parameter $\theta$ scales the transformation.
  * **Specific:** If $\hat{G}$ is energy $\hat{H}$, then $\theta$ is the duration of time t/ℏ because because the Hamiltonian $\hat{H}$ is a specific generator that moves a system through time. Unitary $\hat{U}$ becomes: $e^{-i \hat{H} t / \hbar}$. In this case $\theta = t / \hbar$. But if $\hat{G}$ is momentum $\hat{P}$, then $\theta$ is a displacement in position.
  * **General:** When we use the general form  $\hat{U} = e^{-i \hat{G} \theta}$, we are just saying that $\hat{G}$ is a generator for any kind of transformation (like shifting in space or rotating), and ${\theta}$ is the parameter that tells us "how much" of that transformation to apply.
  * $G$ can then be filled with any linear or non-linear combinations of $\hat{P} + \hat{Q}$ (**Exponentiation**). Hamiltonian $\hat{P}^2 + \hat{Q}^2$ as a quadratic transformation describing total energy, or $q\hat{P} - p\hat{Q}$ as linear, describing displacement as $e^{( -\frac{i}{\hbar} (q\hat{P} - p\hat{Q}))}$
  * Linear Generators $q\hat{P} - p\hat{Q}$ create the Displacement Operators $\hat{D}$. Geometrically, they don't rotate the state around the origin; they slide (translate) it to a new location (q,p) in phase space
  * Quadratic Generators $\hat{Q}^2 + \hat{P}^2$ are Harmonic Oscillator Hamiltonian! These generate actual rotations around the origin of phase space. If you let a state evolve under this generator for a specific "time" or "angle," it's essentially the Quantum Fourier Transform
* <font color="blue">**Conjugate relation: An operator is a Generator of its conjugate variable's translation:** This allows basis change $X$ <> $P$</font>
    * **Clock Z** *represents* Position (Potential) but *generates* Momentum Translation (Kick) - $D_{0,p} \sim Z^p$ applied (moves faster)
    * **Shift X** *represents* Momentum (Kinetic) but *generates* Position Translation (Slide) - $D_{q,0} \sim X^q$ applied (new position)
    * Connected via **Fourier Transform**. Used in $D_{q,p} = \tau^{qp} X^q Z^p$.  Based on commutation relation: $[x̂, p̂] = i ħ$ (for continuous), $ZX = \omega XZ$ (Weyl relation). For example in: Characteristic ($\chi(\xi)$) <> Wigner function.
* <font color="blue">**Kinetic Energy** $\hat{P}^2$ $\rightarrow$ **Momentum** $\hat{P} = \frac{i}{\sqrt{2}}(\hat{a}^\dagger - \hat{a})$ $\rightarrow$ **Shift X operator** $X = e^{-i \hat{P} \delta}$, acts as $|k\rangle \to |k+1\rangle$.</font>
  * Matrix $X$ is Real and Off-Diagonal (Permutation matrix of 0s and 1s). Shift is exponentiated momentum $X \approx e^{-i \hat{P} \delta_x}$
  * Momentum ($\hat{P}$) corresponds to derivative $\approx i(X^\dagger - X)$. (Complex/Imaginary, anti-symmetric).
  * Kinetic Energy ($\hat{P}^2$) corresponds to hopping term/Laplacian $\approx (X + X^\dagger)$. (Real/Symmetric).
  * **Conjugation: Shift X** *represents* Momentum (Kinetic) but *generates* Position Change - $D_{q,0} \sim X^q$ applied (new position)
* <font color="blue">**Potential Energy** $\hat{Q}^2$ $\rightarrow$ **Position** $\hat{Q} = \frac{1}{\sqrt{2}}(\hat{a} + \hat{a}^\dagger)$ $\rightarrow$ **Clock Z operator** $Z = e^{i \hat{Q} \delta}$, acts as $|k\rangle \to \omega^k |k\rangle$ where $\omega=e^{2\pi i/d}$</font>
  * Matrix $Z$ has Complex Eigenvalues (phases on the unit circle) and Diagonal. Operator $\hat{Q}$ has Real eigenvalues (location).
  * Position ($\hat{Q}$) observable has Real eigenvalues (location index $k$). Clock Phase ($Z$) gate is Complex (phases $\omega^k$ on unit circle).
  * **Conjugation: Clock Z** *represents* Position (Potential) but *generates* Momentum (Kick) - $D_{0,p} \sim Z^p$ applied (moves faster)
* <font color="blue">**Unit Circle** (Why Complex Numbers?)</font>
  * [Harmonic oscillator](https://en.wikipedia.org/wiki/Simple_harmonic_motion) moves back and forth: $x(t) = \cos(\omega t)$, described by real numbers: Position ($x$) and Velocity ($v$). Can combined to complex amplitude: $\alpha = x + i p$. Rotation keeps structure (unitary), rotates without changing magnitude (length).
  * Annihilation operator $\hat{a}$ combines Position and Momentum into single complex number: $\hat{a} \propto \hat{Q} + i\hat{P}$. This is $x + iy$ (Euler) $\rightarrow$ Real Axis (Position) $\cos(\omega t)$. Imaginary Axis (Momentum) $i\sin(\omega t)$. Rotation (Time) $e^{i\omega t}$ = Oscillation between $Q$ and $P$
  * Multiplying $\hat{a}$ by $e^{-i\omega t}$ rotates the vector in the complex plane. When the vector is horizontal (Real), the particle has maximum Position $Q$. When the vector is vertical (Imaginary), the particle has maximum Momentum $P$. **90° switch from X to Z = QFT.**
* <font color="blue">**Exponentiation** (explains the quantum operators we use in qubits and qudits)</font>
  * **To get Unitary (transformation) gate, we exponentiate the generator: $\hat{U} = e^{-i \hat{G} \theta}$** (quadratic or higher-order)
  * Generators are constructed from fundamental operators $\hat{Q}, \hat{P}$ (or $\hat{a}$ and $\hat{a}^\dagger$), or $\hat{X}$ and $\hat{Z}$ (modulo 𝑑) for qudit/qubit.
  * Unitary $\hat{U} = e^{-i \hat{H} t / \hbar}$ is defined and computed upon $[\hat{x}, \hat{p}] = i\hbar$ (CCR). Belongs to Weyl-Heisenberg group
    * Example 1 (Displacement): Generator is Linear ($\hat{G} \propto q\hat{P} - p\hat{Q}$) $→$ Creates Shift/Clock gates.
    * Example 2 (Time Evolution): Generator is Quadratic ($\hat{H} \propto \hat{P}^2 + \hat{Q}^2$) $→$  Creates Rotation/Fourier Transform.
* <font color="blue">**Appendix**</font>
  * Mathematically, in QM Time is and Angle. In the expression $e^{-i\theta}$, $\theta$ must be a dimensionless number (an angle in radians). When we write the time-evolution operator as $\hat{U}(t) = e^{-i \hat{H} t / \hbar}$, the quantity $(\hat{H} t / \hbar)$ plays the role of that angle.
  * In QM, states don't move along a trajectory like a baseball but its complex phase rotates. **If a system is in an energy eigenstate (Stationary State) with energy $E$, it stays in that state, but its phase "spins" at a frequency of $\omega = E/\hbar$.**
  * This rotation is also why interference happens: If two parts of a wave function rotate at different speeds (because they have different energies), they will eventually point in opposite directions and cancel out.
  * $\hbar$ (Planck constant in $1.05 \times 10^{-34}$ Joule-seconds) is exchange rate between macroscopic units we use (like Joules and seconds) and internal language of quantum mechanics (phase and frequency). It makes the exponent dimensionless, for example in $\hat{D}(q, p) = e^{-\frac{i}{\hbar} (q\hat{P} - p\hat{Q})}$. This is important because in physics, we can only exponentiate "pure numbers" (dimensionless quantities). Since Energy ($\hat{H}$) is measured in Joules and time ($t$) in seconds, their product is Joule-seconds. We divide by $\hbar$ (also in Joule-seconds) to get dimensionless angle. $\hbar$ defines scale at which quantum effects become dominant.



<font color="blue">***Linear Transformations (≈ First Quantization - Mean Field Theory / Wave Picture)***</font>

Generated by linear or quadratic functions ($\le 2^{\text{nd}}$ degree) of fundamental operators $\hat{Q}, \hat{P}$, or $\hat{X}$ and $\hat{Z}$ (modulo 𝑑). Mathematics: Not orthogonal and form overcomplete basis. Eigenstates of annihilation operator $\hat{a}$.
* <font color="blue">*Hamiltonian $H = \hat{P}^2 + \hat{Q}^2$ (Quadratic). Generates Time Evolution / Rotation which is basis change with Fourier transform*</font>
  * Gaussian: **Phase Shifter** $R(θ)=e^{-i \theta \hat{n}}$ rotates state by angle $θ$. For $θ=π/2$ (quarter turn), 90° = Fourier Transform.
    * Generating operator is number operator $\hat{n}$ with Hamiltonian $H \propto \hat{Q}^2 + \hat{P}^2$ with rotation operator $R(θ)=e^{-i \theta (Q^2 + P^2)}$
    * $\hat{Q}$-$\hat{P}$ swap represents how potential and kinetic energy switch back and forth forever = basis change (rotation)
    *  If we wait a quarter-period ($\theta = \pi/2$), we have physically performed a Fourier Transform.
  * Clifford (Qudit): **Quantum Fourier Transform (QFT)** is exactly Quarter Turn π/2 to rotate $Z$-eigenbasis (Position) into $X$-eigenbasis (Momentum) due to conjugate relation. Basis change by action on computational states $\text{QFT}*d |j \rangle$ =  ${0}^{d-1}  \omega^{jk} |k \rangle$.
    * Basis change: maps single, localized basis state $|j\rangle$ into uniform superposition of all basis states tagged with phase $\omega^{jk}$.
    * $W X W^\dagger = Z$. Quantum Fourier Transform turns Shift into Clock $\rightarrow$ QFT diagonalizes Shift $X$ operator
    * $W Z W^\dagger = X^\dagger$. Inverse Quantum Fourier Transform. QFT turns Clock into inverse Shift $\rightarrow$ Qudit $Z$ operator applies the phase $\omega^j = e^{2\pi i j / d}$ based on the *single* qudit index $j$.
  * Clifford (QUBIT): **Hadamard ($\hat{H}$)**. Conjugate relation allows basis change $Z$<>$X$  via equal superposition $H|0\rangle = |+\rangle$.
* <font color="blue">[Displacement Operator](https://en.wikipedia.org/wiki/Displacement_operator) *$G = q\hat{P} - p\hat{Q}$ (Linear). Generates shifts in [Optical Phase Space](https://en.wikipedia.org/wiki/Optical_phase_space)*</font>
  * Hamiltonians linear in $\hat{Q}, \hat{P}$ (degree 1) generate **Displacements**.
    * $H \propto \hat{P} \rightarrow$ generates shift in $Q$. And $H \propto \hat{Q} \rightarrow$ generates shift in $P$.
    * Computation via Heisenberg-Weyl Group with of all possible shifts and phases ($X^q Z^p$).
    * Phase factor $e^{i\pi qp/d}$ required because $X$ and $Z$ do not commute (Aharonov-Bohm effect in phase space).
  * Gaussian: **Displacement** $\hat{D}(q, p) = e^{( -\frac{i}{\hbar} (q\hat{P} - p\hat{Q})}$ with generator $H  \propto (q\hat{P} - p\hat{Q})$
  * Clifford (Qudit): **Displacement** $D_{q,p} = e^{i\pi qp/d} X^q Z^p$ (combines shift $q$ units and phase $p$ units)
* <font color="blue">*Kinetic $\rightarrow$ Momentum $\rightarrow$ Shift $\rightarrow$ Bit Flip/Pauli X $\rightarrow$ Linear Generator $\hat{P}$*</font>
  * Gaussian: **Momentum** $\hat{P} = \frac{i}{\sqrt{2}}(\hat{a}^\dagger - \hat{a})$. Represents momentum, but generates position shift:
  * Clifford (Qudit): **Shift**  $\hat{X}^a |j\rangle = |j+1\rangle \pmod d\rangle$. |0⟩→|1⟩→|2⟩→|d−1⟩→|0⟩. Eigenvalues of $X$ are powers of $\omega$ (via Fourier). Because:  In "Momentum Basis", Shift operator is diagonal! Looks like Clock matrix.Because $X = \text{DFT}^\dagger \cdot Z \cdot \text{DFT}$.
  * Clifford (Qubit): **Pauli X**: $\sigma_x |j\rangle = |j+1 \pmod 2\rangle$. Swaps 0<>1. $ZX = -XZ$. Phase is ω = $e^{i 2\pi/2}$ = -1, Eigenvalues 1, -1.
* <font color="blue">*Potential $\rightarrow$ Position $\rightarrow$ Clock $\rightarrow$ Phase Flip/Pauli-Z $\rightarrow$ Linear Generator $\hat{Q}$*</font>
  * Gaussian: **Position** $\hat{Q} = \frac{1}{\sqrt{2}}(\hat{a} + \hat{a}^\dagger)$ is observable. Position generates translation in Momentum:
  * Clifford (Qudit): **Clock** $\hat{Z}^b |k\rangle = \omega^k |k\rangle$ where $\omega=e^{2\pi i/d}$ phase gradient + primitive $d$-th root of unity (=Eigenvalues). ($e^{ix}$ is the clock gate with complex eigenvalues on the unit circle) Raised to $d$-th power ($\omega^0, \omega^1, \dots, \omega^{d-1}$) to generate $d$ solutions to tag each basis state k. For $d=3$, $\omega = e^{i 2\pi/3}$  (unitary but not Hermitian).
  * Clifford (Qubit): **Phase Flip (Pauli Z)**. $d=2$ primitive root of unity: $\omega_2 = e^{2\pi i / 2} = e^{i\pi} = -1$ $→$  $\sigma_z |j\rangle = (-1)^j |j\rangle = \omega^j |j\rangle$. This is mod 2 version: $ZX = -XZ (ω = -1)$. Adds a $-1$ phase to $\sigma_z |1\rangle = -|1\rangle$, but leaves $\sigma_z |0\rangle = |0\rangle$. = Pauli $Z$ matrix is both Unitary (preserves probabilities) and Hermitian (is its own conjugate). We can observe it directly.
* <font color="blue">*Gaussian: Pauli-Y*</font>
  * is just a product, not independent operator! In qubit $\sigma_y = i \sigma_x \sigma_z$ is just combination of bit and phase flip. For qudits: family of Generalized Pauli Operators defined by product of shifts and clocks: $P_{q,p} = X^q Z^p$.
  * **For $d=2$ (Qubits):** $X^1 Z^0 = X$ and $X^0 Z^1 = Z$ and $X^1 Z^1 \propto Y$ ($Y$ is just $(1,1)$ point on the grid).
  * **For $d=3$ (Qutrits):** $X Z$, $X Z^2$, $X^2 Z$, etc. None of these is uniquely called "$Y$". "Pauli $Y$" doesn't vanish; it just becomes one of the many $D_{q,p}$ displacement operators ($q \neq 0, p \neq 0$)
* <font color="blue">*Shear/Phase Gate*</font>
  * Gaussian: **Quadratic Phase (Shear)**: Generator $H \propto \hat{Q}^2$. Hamiltonian $H \propto \hat{Q}^2$ (a "Kerr" nonlinearity or a Shear) preserves Position $Q$ but shears Momentum $P \to P + Q$.
  * Clifford (Qubit): **Phase Gate ($\hat{S}$ or $\hat{P}$)**. $Z \to Z$, $X \to Y \sim XZ$. Phase gate is diagonal. $S = \text{diag}(1, i, \dots)$. It adds a quadratic phase $k^2$ to the wavefunction.
* <font color="blue">*Entanglement*</font>
  * Gaussian: **Squeezer + Beam Splitter = Entanglement** $\approx$ CNOT.  Squeeze state to ellipse, then rotate ellipse 45° via Beam Splitter $\rightarrow$ noise becomes correlated between two axes = this is Entanglement.
    * **Squeezer** is $\hat{S}(r) = e^{\left( -\frac{i r}{2\hbar} (\hat{Q}^2 - \hat{P}^2)\right)}$ with generator $H \propto (\hat{Q}^2 - \hat{P}^2)$ $= \hat{Q}\hat{P} + \hat{P}\hat{Q}$. Scales axes: stretches $Q$ and shrinks $P$ (or vice versa). Multiplies position by $e^{-r}$ and momentum by $e^r$.
    * **Beam Splitter** $\hat{B}(\theta) = e^{-i \theta (\hat{Q}_1 \hat{P}_2 - \hat{Q}_2 \hat{P}_1)}$ with generator $H \propto \hat{Q}_1\hat{P}_2 - \hat{Q}_2\hat{P}_1$. This is the angular momentum operator in phase space. It performs passive rotation between two modes.If $\theta = \pi/4$ (50:50 Beam Splitter), it mixes the two inputs equally:$\hat{a}_1 \to \frac{\hat{a}_1 + \hat{a}_2}{\sqrt{2}}$ and $\hat{a}_2 \to \frac{\hat{a}_1 - \hat{a}_2}{\sqrt{2}}$.
  * Clifford (Qudit): **C-SUM ($CX$)** $|c\rangle|t\rangle \to |c\rangle|t \oplus c\rangle  \pmod d\rangle$.  Adds control bit to target bit modulo d.
    * $\hat{C}_{\text{SUM}} = e^{-i \hat{Q}_1 \hat{P}_2}$ with generator $\hat{Q}_1 \hat{P}_2$. Couples Position particle 1 to Momentum (shift) particle 2.
    * C-SUM gate built via: Beam Splitter + Squeezing + Inverse Beam Splitter
  * Clifford (Qubit): **CNOT**. $|c\rangle|t\rangle \to |c\rangle|t \oplus c\rangle  \pmod 2\rangle$. Adds control bit to target bit modulo 2.

<font color="blue">***Non-Linear Transformations (≈ Second Quantization - Many-Body Interaction)***</font>

Generated by cubic or higher functions ($\ge 3^{\text{rd}}$ degree) of fundamental operators $\hat{Q}, \hat{P}$, or $\hat{X}$ and $\hat{Z}$ (modulo 𝑑). Break symmetry. Mathematics: Form orthonormal basis. Eigenstates of number operator $\hat{n} = \hat{a}^\dagger \hat{a}$.
* <font color="blue">*Cubic Phase*</font>
  * Non-Gaussian: **Cubic Phase** $\hat{U} = e^{i\gamma \hat{Q}^3}$ with generator $H \propto \hat{Q}^3$.  Cubic Phase gate $V(\gamma) = e^{i \gamma \hat{Q}^3}$ distorts phase space. A circle (coherent state) becomes a "banana" shape with negative Wigner function regions—the signature of non-classicality.
  * Non-Clifford (Qudit): **Discrete Cubic Phase ($\hat{T}_d$)** (mod d). Generalized Magic Gates, often diagonal phases like $e^{i \frac{2\pi}{d^3} k^3}$. Corresponds to degree-3 polynomial phases: $T_d |k\rangle = \omega^{k^3} |k\rangle$. Exactly analogous to the cubic phase gate: $V(\gamma) = e^{i\gamma \hat{x}^3}$.  
  * Non-Clifford (QUBIT): **$\hat{T}$-gate** (magic gate) degree-3 (or $\pi/8$ gate): $\hat{T} = e^{-i \pi/8 \hat{Z}}$ - is cubic phase in discrete phase space (mod 2) : $\pi/8$ phase. The Clifford group maps Pauli operators to Pauli operators (Level 2).
* <font color="blue">*Other Operators*</font>
  * Non-Gaussian: **Kerr** $\hat{U} = e^{i\chi \hat{n}^2}$ where $\hat{n} = \hat{a}^\dagger \hat{a}$. Generator $\propto (\hat{Q}^2 + \hat{P}^2)^2$ (quartic). Since $\hat{n} \sim Q^2 + P^2$, then $\hat{n}^2 \sim (Q^2 + P^2)^2 \sim Q^4 + \dots$This is a Quartic (4th order) potential. This represents a medium where the refractive index changes based on intensity. In computation, it is often used to build "Cat States" (macroscopic superpositions), which are excellent candidates for error-corrected logical qubits (Bosonic Codes).
  * Non-Clifford (Qudit): **General high-order phase gates**
  * Non-Clifford (QUBIT): **Toffoli Gate** (also Non-Clifford and Universal with $\hat{H}$ and $\hat{T}$)

<font color="blue">**Schrödinger Equation**

*Description of Quantum Systems: Classical vs Quantum Methods*

Quantum systems in chemistry and physics can be described and computed with either classical and quantum methods:

* **Model**: Classical (Newton Mechanics) vs Quantum (Schrödinger equation)
* **Type**: Static (State) vs Dynamical (Evolution)
* **Computing**: Classical Computing vs Quantum Computing

<font color="blue">*1. Classical (Newtonian) Mechanics (**Static and Dynamic**): $F = m \cdot a$*
* Ignores electrons and consider atoms as spheres connected by springs (force fields)
* **Classical - Dynamic**: Molecular dynamics (MD). How proteins enfold and move? GROMACS, NAMD, AMBER.
* **Classical - Static**: Molecular Docking / Energy Minimization. Geometrically: Does medication fit to enzyme? AutoDock, Rosetta.
* **Quantum computing methods:** Future: Navier Stokes (fluid dynamics) with HHL to solve linear systems and differential equations exponentially faster than classical computers. For weather and climate: simulate weather with a much finer grid (e.g. 100m)

<font color="blue">*2. Quantum Mechanics (**Static**): Time-independent Schrödinger Equation $\hat{H}|\psi\rangle = E|\psi\rangle$*
* Includes electrons. **Approximates Ground State via optimization** with variation $\delta \langle H \rangle = 0$ to find global optimum in energy landscape. Stationary state important for calculating binding energies: If $\psi$ is an eigenstate of $\hat{H}$, then time evolution is trivial: $\Psi(t) = \psi e^{-iEt/\hbar}$. Probability density $|\Psi(t)|^2$ remains constant over time = states are stationary.
* **Classical computing:** Hartree-Fock (mean field of $\psi$) but ignore correlation energies, DFT (electron density $\rho$) approximates correlation energies with functionals (but require correct functional assumed), Post-HF (Coupled cluster, configuration interaction) calculate correlation energies but only small systems since their complexity grows exponentially
* **Quantum computing** (**NISQ**): Variational Quantum Eigensolver. Quantum Advantage: finds correlation energy with entanglement.
  * In classical quantum chemistry, correlation energies are either ignored (HF), approximated (DFT), or, while theoretically calculable exactly, are only feasible for small quantum systems (post-Hartree-Fock methods like coupled cluster or configuration interaction).
  * For classical computers, complexity grows exponentially, this is why quantum computers will play a significant role in analyzing larger, more complex systems.

<font color="blue">*3. Quantum Mechanics (**Dynamic**): Time-dependent Schrödinger Equation $i\hbar \frac{\partial}{\partial t} \Psi(\mathbf{r}, t) = \hat{H} \Psi(\mathbf{r}, t)$*
* **Simulate Time Evolution** $e^{-i\hat{H}t}$ of Quantum State $|\Psi(t)\rangle = e^{-i\hat{H}t/\hbar} |\Psi(0)\rangle$ via Unitary Transformation. For Quantum Chaos, OTOC, Scrambling (Black Holes) reaction dynamics (observe how bonds break and reform during a collision). No forward theorem!
* **Classical computing**: Time-Dependent DFT (TD-DFT) used to classically calculate excitations (fluorescence, spectra). However only for short times and small systems. To solve the time-dependent equation $i\hbar \frac{\partial}{\partial t} \Psi = \hat{H} \Psi$ classically at scale, you have to apply operator $e^{-i\hat{H}t/\hbar}$ to wavefunction. In system with $N$ particles, complexity grows exponentially with $N$.
* **Quantum computing** (**Fault-Tolerant**): Central challenge in quantum dynamics (Simulating Quantum Time Evolution):
  * Molecule evolves under all its forces (Kinetic + Potential) simultaneously, but QC can only do one. But unitary propagator $U(t) = e^{-i\hat{H}t}$ is implemented on quantum hardware that only supports a discrete set of elementary gates at once.
  * Three approaches: LCU, Trotterization, Qubitization (with Quantum Walks and Block Encoding). And with the OTOC metric to measure information scrambling during quantum evolution (https://research.google/blog/a-verifiable-quantum-advantage/)


<font color="blue">*Appendix: Why we can solve the Schrödinger equation analytically only for very small systems*

* **Analytical** solution for Schrödinger in Hydrogen atom. But a second electron add electron-electron repulsion = three-body problem.
* **Approximate** Schrödinger with DFT and HF on classical computers via **optimization (variation)** for larger systems than Hydrogen
* **Simulate** time evolution for even larger systems = Quantum computer doesn't store information anymore - **it is the Hilbert space!**

```python
# @title Trotterization (Suzuki-Trotter)
# ==========================================
# TROTTERIZATION (Suzuki-Trotter Decomposition)
# ==========================================
from IPython.display import display, Markdown


print(40*"=")
print("TROTTERIZATION (Suzuki-Trotter)")
print(40*"=")

display(Markdown(r"* **Trotter (Product) Approximation**: $e^{-i(H_{int} + H_{field})t} \approx \left( e^{-i H_{int} \Delta t} \cdot e^{-i H_{field} \Delta t} \right)^n$ (in each step it switches between forces)"))
display(Markdown(r"* Approximates propagator by decomposing Hamiltonian $\hat{H} = \sum \hat{h}_j$ into small time steps $\Delta t = t/n$."))
display(Markdown(r"* Trotterization: product decomposition: e.g. apply OTOC by breaking $e^{-iHt}$ into thousands of small gates to approximate it."))
display(Markdown(r"* **$X$-Rotation (Transverse Field $X$:** flip spins (Dynamics) Term $e^{-i X \theta}$ corresponds exactly to an RX gate."))
display(Markdown(r"* **$ZZ$-Interaction $Z_0 Z_1$:** align/anti-align spins (Entanglement). $e^{-i Z \otimes Z \theta}$ operator not natively on QC. Constructed  with sequence: CNOT $\rightarrow$ RZ $\rightarrow$ CNOT."))

# 1. GENERAL PHYSICS ENGINE

# Generates single time-slice (Trotter step). Simulates Hamiltonian: H = J * (Z0 Z1) + B * (X0 + X1)
def generate_trotter_step(qubits, dt, J, B):
    """
    qubits: List of qubits to simulate.
    dt: size of time step.
    J: Strength of Interaction (Entanglement).
    B: Strength of Transverse Magnetic Field (Dynamics).
    """
    n = len(qubits)

    # --- Part A: Interaction Terms (ZZ) - "Entangler" ---
    """
    Target Physics: exp(-i * J * Z * Z * dt)
    Problem: Quantum computers don't have a native "ZZ" gate.
    Solution: synthesize it using a "Gadget": CNOT -> Rz -> CNOT.
    Logic: If J=0, the qubits don't interact, so skip these gates
    """
    if J != 0:
        for i in range(n - 1):
            # 1. Compute Parity (store result in second qubit)
            yield cirq.CNOT(qubits[i], qubits[i+1])

            # 2. Add Phase (proportional to interaction energy J)
            # Rz(phi) = exp(-i * Z * phi/2). Want exp(-i * J * dt * Z). So phi/2 = J*dt  ->  phi = 2*J*dt
            yield cirq.rz(rads=2 * J * dt).on(qubits[i+1])

            # 3. Uncompute Parity (restore basis)
            yield cirq.CNOT(qubits[i], qubits[i+1])

    # --- Part B: Transverse Field Terms (X) - The "Rotator" ---
      # Target Physics: exp(-i * B * X * dt)
      # corresponds exactly to a Rotation around X-axis (Rx gate).
      # Logic: If B=0, the magnetic field is off, skip rotation.
    if B != 0:
        for i in range(n):
            # Rx(theta) = exp(-i * X * theta/2). Want exp(-i * B * dt * X). So theta = 2*B*dt.
            yield cirq.rx(rads=2 * B * dt).on(qubits[i])

# Hyperparameters
num_qubits = 2
t_max=2.5 # simulation_time
steps=30 # trotter steps

# Simulates the system from t=0 to t_max. Returns probabilities of finding system in |00> or |11>
def get_probabilities_over_time(J, B, t_max=t_max, steps=steps):
    qubits = cirq.LineQubit.range(num_qubits)
    time_points = np.linspace(0, t_max, steps)
    probs_00 = []
    probs_11 = []
    print(f"Running Simulation for J={J}, B={B}...")

    for t in time_points:
        circuit = cirq.Circuit()

        # Build Time Evolution Circuit. Divide total time 't' into n smaller Trotter steps
        # to ensure approximation holds (dt must be small).
        if t > 0:
            dt = t / 10
            for _ in range(10):
                circuit.append(generate_trotter_step(qubits, dt, J, B))

        # --- Simulation ---
        simulator = cirq.Simulator()

        # Explicitly provide 'qubit_order' to force a 2-qubit simulation
        # even if circuit is empty (at t=0) or has disconnected qubits.
        result = simulator.simulate(circuit, qubit_order=qubits)

        # --- Measurement ---
        # result.final_state_vector returns amplitudes [a00, a01, a10, a11]. Probability = |amplitude|^2
        probs_00.append(np.abs(result.final_state_vector[0])**2) # Index 0 is |00>
        probs_11.append(np.abs(result.final_state_vector[3])**2) # Index 3 is |11>

    return time_points, probs_00, probs_11

# 2. VISUALIZE CIRCUIT
print(40*"=")
print("Circuit Logic Verification")
print(40*"=")
# Demo 1: Full Hamiltonian (Standard Case). Expectation: CNOTs are present (Interaction is ON)
qubits_demo = cirq.LineQubit.range(num_qubits)
circuit_full = cirq.Circuit(generate_trotter_step(qubits_demo, dt=0.1, J=1.0, B=1.0))
print("\n(a) Structure for J=1, B=1 (Standard - Full Interaction):")
print(circuit_full)

# Demo 2: Field Only. Expectation: CNOTs are MISSING (Interaction is OFF). Circuit is smaller.
circuit_field = cirq.Circuit(generate_trotter_step(qubits_demo, dt=0.1, J=0.0, B=1.0))
print("\n(b) Structure for J=0, B=1 (Field Only - No CNOTs):")
print(circuit_field)

print(40*"=")
display(Markdown(r"Simulation (Ising) $H = Z_0 Z_1 + X_0 + X_1$. Generalized: $H = J \cdot (Z_0 Z_1) + B \cdot (X_0 + X_1)$"))
print(40*"=")
# 3. SIMULATION RESULTS
plt.figure(figsize=(10, 4))

# === Subplot 1: Detailed Dynamics (Balanced Forces) ===
  # The Field tries to flip qubits, the Interaction tries to align them.
  # Result: Complex oscillation (Superposition).
plt.subplot(1, 2, 1)
times, p00, p11 = get_probabilities_over_time(J=1.0, B=1.0)
plt.plot(times, p00, label='P(|00>) - Start', color='blue', linestyle='--')
plt.plot(times, p11, label='P(|11>) - Flipped', color='red')
plt.title("Scenario 1: Standard Dynamics (J=1, B=1), \nCompetition between Field and Interaction")
plt.xlabel("Time (t)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True, alpha=0.3)

# === Subplot 2: Parameter Sweep (Comparison) ===
plt.subplot(1, 2, 2)
# Case A: Field Only (J=0): Pure Rabi Oscillations. Qubits rotate cleanly |0> -> |1> -> |0>.
t_a, p00_a, _ = get_probabilities_over_time(J=0.0, B=1.0)
plt.plot(t_a, p00_a, label='Field Only (J=0)', color='blue', linestyle='--')

# Case B: Strong Interaction (J=5): Interaction Blockade / Quantum Zeno Effect.
# energy cost to flip is too high. system is "frozen" in |00>
t_b, p00_b, _ = get_probabilities_over_time(J=5.0, B=1.0)
plt.plot(t_b, p00_b, label='Strong Interaction (J=5)', color='red')
plt.title("Scenario 2: Interaction Blockade Effect\nStrong J prevents Bit-Flips")
plt.xlabel("Time (t)")
plt.ylabel("Probability P(|00>)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# @title Qubitization (Quantum Walk, Block Encoding, LCU)
# ==========================================
# OTOC (Quantum Implementation)
# ==========================================
from IPython.display import display, Markdown

print(40*"=")
print("Qubitization (Quantum Walk, Block Encoding, LCU)")
print(40*"=")

display(Markdown(r"* Qubitization is an algebraic trick. It transforms the problem of walking through time into a problem of walking through the eigenvalues of an operator."))
display(Markdown(r"* **LCU Decomposition:** Approximate operator by decomposing propagator $e^{-iHt}$ via Taylor series in weightes sum of unitaries $H = \sum_{l=0}^{L-1} \alpha_l U_l$. (non-unitary, represents energy). Then **Normalization:** $\lambda = \sum |\alpha_k|$ $\rightarrow$ $H/\lambda$. If $H=$ Dirac $\mathcal{D}$ = as benchmark, its algebraic structure (sum of Paulis) efficient."))
display(Markdown(r"* **LCU Oracles**: 1. ***PREPARE Oracle*** ($P$): encodes coefficients $\alpha_j$ into amplitudes of ancilla. Output is state where probability of measuring state $j$ is proportional to $\alpha_j$: $P|0\rangle = \frac{1}{\sqrt{\|\alpha\|_1}} \sum_{j=0}^{m-1} \sqrt{\alpha_j} |j\rangle$. ***2. SELECT Oracle*** ($S$) apply unitaries $U_l$ to target register state, conditioned on state of ancilla (entangles). If ancilla is state $|j\rangle$, it applies $U_j$ to target state $|\psi\rangle$. $S = \sum_{j=0}^{m-1} |j\rangle\langle j| \otimes U_j$ ***3. Un-prepare***: Apply inverse $P^\dagger$ to ancilla. ***4. Measure***: ancilla."))
display(Markdown(r"* **Block Encoding $U_H$:** LCU embeds $\hat{H}/\lambda$ in top-left corner of quantum walk $U_H = \begin{pmatrix} \hat{H}/\lambda & \dots \\ \dots & \dots \end{pmatrix}$. If Ancilla is $|0\rangle$, we apply $H$. If Ancilla is $|1\rangle$, apply junk."))
display(Markdown(r"* **Quantum Walk** $Q$: add reflections to $U$ turns it into walk that moves through energy spectrum. By reflecting junk back into system ($Q = R \cdot U$), a rotation is created. Eigenvalues of $Q$ are linked to eigenvalues $E$ of $H$ via $\mu = e^{\pm i \arccos(E/\lambda)}$ (energy mapping). Physics is in angle (Isomorphism): $E_k = \lambda \cos(\theta_k)$. QW transforms *energy information* (scalar magnitude) into *phase information* (how fast rotates). High Energy $\rightarrow$ Small Angle. Zero Energy $\rightarrow$ $\cos(90^\circ)=0$."))

print(40*"=")
print("Simulation: Qubitization for OTOC")
print(40*"=")

display(Markdown(r"* Example: OTOC (scrambling of black hole or complex molecule). Time Reversal in Qubitization: invert Reflection and Oracles. **Shows that TIME in Qubitization is actually an ANGLE in Hilbert space**."))
display(Markdown(r"* Scrambling: Even with 1 system qubit, measurement changes if Forward and Backward walks don't perfectly cancel out because of the $X$ gate (butterfly). Code is gate-based with no god-mode access to $e^{-iHt}$."))
print("\n")

# --- 1. Define the Oracles (Same as before) ---

class PrepareOracle(cirq.Gate):
    def __init__(self, coeffs):
        super(PrepareOracle, self).__init__()
        self.coeffs = coeffs
        self.l1 = sum(coeffs)
        self.theta = 2 * np.arccos(np.sqrt(coeffs[0] / self.l1))
    def _num_qubits_(self): return 1
    def _decompose_(self, qubits):
        yield cirq.ry(self.theta).on(qubits[0])

class SelectOracle(cirq.Gate):
    def _num_qubits_(self): return 2
    def _decompose_(self, qubits):
        anc, sys = qubits
        yield cirq.X(sys).controlled_by(anc, control_values=[0])
        yield cirq.Z(sys).controlled_by(anc, control_values=[1])

# --- 2. Define the Qubitization Walk Step (The Engine) ---

class QubitizationWalkStep(cirq.Gate):
    """A single step of the Quantum Walk Q."""
    def __init__(self, coeffs):
        super(QubitizationWalkStep, self).__init__()
        self.prep = PrepareOracle(coeffs)
        self.sel = SelectOracle()

    def _num_qubits_(self): return 2

    def _decompose_(self, qubits):
        anc, sys = qubits
        # Q = Reflection * LCU_Operator
        # Part 1: LCU Operator (Block Encoding)
        yield self.prep.on(anc)
        yield self.sel.on(anc, sys)
        yield cirq.inverse(self.prep).on(anc)
        # Part 2: Reflection on Ancilla
        yield cirq.Z(anc)

# --- 3. The OTOC Experiment as example use case for qubitization ---

def run_qubitization_otoc(steps=2):
    q_anc = cirq.NamedQubit("Ancilla")
    q_sys = cirq.NamedQubit("System")
    circuit = cirq.Circuit()

    walk_step = QubitizationWalkStep(coeffs=[0.6, 0.8])

    # STEP A: Forward Evolution (Forward Walk)
    for _ in range(steps):
        circuit.append(walk_step.on(q_anc, q_sys))

    # STEP B: The Butterfly Perturbation (V)
    # We apply X to the system
    circuit.append(cirq.X(q_sys))

    # STEP C: Backward Evolution (Inverse Walk)
    # This is the 'Time Reversal' unique to OTOCs
    for _ in range(steps):
        circuit.append(cirq.inverse(walk_step).on(q_anc, q_sys))

    # STEP D: Measurement (W)
    circuit.append(cirq.measure(q_sys, key='otoc_result'))

    print(f"--- Qubitization OTOC Circuit ({steps} steps) ---")
    print(circuit)

    simulator = cirq.Simulator()
    result = simulator.run(circuit, repetitions=1000)
    print("\nResults (Histogram):")
    print(result.histogram(key='otoc_result'))

run_qubitization_otoc(steps=4)
```

```python
# @title Out-of-Time-Order Correlators (OTOCs)

# OTOC (Quantum Dynamics, Quantum Chaos and Scrambling)
# ==========================================
# OTOC (Linear Algebra Implementation)
# ==========================================

from scipy.linalg import norm
from IPython.display import display, Markdown

print(40*"=")
print("OTOC (Quantum Dynamics, Quantum Chaos and Scrambling)")
print(40*"=")

display(Markdown(r"* **OTOC** messen das Scrambling von Information. Wenn du einen Operator W(t) zu einer Zeit t hast und einen Operator V(0) zum Zeitpunkt 0, misst der OTOC $ \langle W^\dagger(t) V^\dagger(0) W(t) V(0) \rangle $, **wie stark diese Operatoren nach der Zeit t nicht mehr kommutieren**. QM Äquivalent zum Schmetterlingseffekt."))
display(Markdown(r"* **OTOC** $C(t) = \langle [\hat{W}(t), \hat{V}(0)]^\dagger [\hat{W}(t), \hat{V}(0)] \rangle$ is metric to quantify quantum chaos and rate of scrambling (quantum Lyapunov exponent). **Operator Spreading:** local operator $\hat{V}$ (Pauli gate) evolves under chaotic Hamiltonian into non-local operator $\hat{V}(t) = e^{iHt}\hat{V}e^{-iHt}$ (Eigenstate Thermalization)"))
display(Markdown(r"* OTOC is calculated via sequence: ***Forward Evolution***: Apply Qubitization Walk $n$ times (simulating time $+t$). ***Perturbation***: Apply Butterfly operator $\hat{V}$ ($X$ gate). ***Backward Evolution***: Apply inverse Qubitization Walk $n$ times (simulating time $-t$). ***Measurement***: Apply probe operator $\hat{W}$ and check correlation."))
display(Markdown(r"* **Four-Point Function:** evolve forward and backward in time: $W(t) = e^{iHt} W(0) e^{-iHt}$. Schrödinger (moving states) $\rightarrow$ Heisenberg (moving operators)"))
display(Markdown(r"* **Time-Dependent Schrödinger** $|\Psi(t)\rangle = e^{-iHt} |\Psi(0)\rangle$ tells how state moves. OTOC describes if chaotic or ordered. Time-Dependent part hidden in $\hat{W}(t)$."))
display(Markdown(r"* **Heisenberg Evolution** $\hat{W}(t) = e^{iHt} \hat{W}(0) e^{-iHt}$ to calculate $\hat{W}(t)$, where $e^{-iHt}$ is solution to Schrödinger. Without Schrödinger, no $\hat{W}(t)$ and no OTOC."))
display(Markdown(r"* **Loschmidt Echo**: measure OTOC on QC: 1. Evolve forward ($e^{-iHt}$) 2. Perturbation ($\hat{V}$) 3. Evolve backward ($e^{+iHt}$ - time reversal) 4. Measure overlap"))
display(Markdown(r""))

def get_unitary_from_circuit(qubits, t, h_x, h_z, steps=25):
    """
    Creates the Unitary Matrix U(t) for a specific Hamiltonian setup.
    H = Sum(Zi Zj) + h_x * Sum(Xi) + h_z * Sum(Zi)
    """
    circuit = cirq.Circuit()
    dt = t / steps
    n = len(qubits)

    for _ in range(steps):
        # 1. Interaction (ZZ) - The "Scrambler"
        for i in range(n - 1):
            circuit.append(cirq.CNOT(qubits[i], qubits[i+1]))
            circuit.append(cirq.rz(rads=2 * dt).on(qubits[i+1]))
            circuit.append(cirq.CNOT(qubits[i], qubits[i+1]))

        # 2. Transverse Field (X) - The "Driver"
        if h_x != 0:
            for i in range(n):
                circuit.append(cirq.rx(rads=2 * h_x * dt).on(qubits[i]))

        # 3. Longitudinal Field (Z) - The "Chaos Generator"
        # Adding this term breaks integrability and causes chaos.
        if h_z != 0:
            for i in range(n):
                circuit.append(cirq.rz(rads=2 * h_z * dt).on(qubits[i]))

    # Extract the full unitary matrix (2^N x 2^N)
    return cirq.unitary(circuit)

def calculate_otoc_curve(num_qubits, h_x, h_z, time_points):
    """
    Calculates the OTOC C(t) = || [W(t), V] ||^2
    V = X on Qubit 0 (The Perturbation)
    W = Z on Qubit N-1 (The Probe)
    """
    qubits = cirq.LineQubit.range(num_qubits)
    otoc_values = []

    # Define Operators V (at t=0) and W (at t=0)
    # We build them as full matrices (2^N x 2^N)

    # V = X_0 (Tensor product: X I I I...)
    op_list_V = [np.eye(2)] * num_qubits
    op_list_V[0] = np.array([[0, 1], [1, 0]]) # X
    V = op_list_V[0]
    for i in range(1, num_qubits):
        V = np.kron(V, op_list_V[i])

    # W = Z_last (Tensor product: ...I I I Z)
    op_list_W = [np.eye(2)] * num_qubits
    op_list_W[-1] = np.array([[1, 0], [0, -1]]) # Z
    W = op_list_W[-1]
    for i in range(num_qubits-2, -1, -1):
        W = np.kron(op_list_W[i], W)

    #print(f"Simulating: h_x={h_x}, h_z={h_z}...")

    for t in time_points:
        if t == 0:
            otoc_values.append(0)
            continue

        # 1. Get Time Evolution Operator U(t)
        U = get_unitary_from_circuit(qubits, t, h_x, h_z)
        U_dag = U.conj().T

        # 2. Evolve W forward in time (Heisenberg Picture)
        # W(t) = U^dagger * W * U
        W_t = U_dag @ W @ U

        # 3. Calculate Commutator C = [W(t), V] = W(t)V - VW(t)
        commutator = W_t @ V - V @ W_t

        # 4. Measure "Size" of commutator (Frobenius Norm)
        # This tells us how non-commutative they have become
        c_val = np.linalg.norm(commutator) ** 2
        otoc_values.append(c_val)

    return otoc_values

# --- Simulation Setup ---
n_qubits = 5  # Small enough for matrix multiplication
times = np.linspace(0, 15.0, 30)

# Scenario 1: Ordered / Integrable (Ising Model)
# No longitudinal field (h_z = 0)
otoc_ordered = calculate_otoc_curve(n_qubits, h_x=1.0, h_z=0.0, time_points=times)

# Scenario 2: Chaotic (Non-Integrable)
# We add a longitudinal field (h_z = 0.5) to break symmetry
otoc_chaotic = calculate_otoc_curve(n_qubits, h_x=1.0, h_z=0.5, time_points=times)

# --- Plotting ---
plt.figure(figsize=(6, 3))
small_size = 8 # font size
plt.plot(times, otoc_ordered, label='Ordered (h_z=0)', color='blue', linestyle='--')
plt.plot(times, otoc_chaotic, label='Chaotic (h_z=0.5)', color='red', linewidth=2)
plt.title(f'Quantum Scrambling (with Qubitization): OTOC Growth (N={n_qubits} Qubits)', fontsize=small_size + 1)
plt.xlabel('Time (t)', fontsize=small_size)
plt.ylabel('OTOC Commutator Magnitude', fontsize=small_size)
plt.legend(fontsize=small_size - 1) # fint size on legend slightly smaller
plt.tick_params(axis='both', labelsize=small_size)

plt.grid(True, alpha=0.3)
plt.tight_layout() # Ensures labels don't get cut off at smaller sizes
plt.show()

display(Markdown(r"* **Maldacena-Shenker-Stanford Schranke**: Vermutung, dass Schwarze Löcher die schnellsten Scrambler der Natur sind"))
display(Markdown(r"* **Quantum Butterfly Effect**: Calculates OTOC between first qubit and other qubits. Proof how information spreads physically (light cone of scrambling in real)."))
display(Markdown(r"* **Lieb-Robinson Light Cone (Bound)** (diagonal V-shape): information has speed limit **Butterfly Velocity** $v_B$ (slope of diagonal line) as it travels down chain"))
display(Markdown(r"* **Poincare Recurrence** (Left) with Ballistic Echoes. Information remains particle-like and bounces back coherently to original state."))
display(Markdown(r"* **Quantum Chaos** (Right). Information scrambles (thermalizes). By t=20, system is perfect model for how information is hidden in Black Hole."))
display(Markdown(r"* **Interaction strength $J$** increase makes slope steeper (information moves faster) and quantum chaos $h_z$ grows faster. Light Cone of Information Scrambling:"))

def calculate_otoc_heatmap(num_qubits, h_x, h_z, t_max, t_steps):
    """
    Calculates OTOC for all qubits over time to visualize the Light Cone.
    """
    qubits = cirq.LineQubit.range(num_qubits)
    times = np.linspace(0, t_max, t_steps)
    # Matrix to store results: Rows = Qubits, Cols = Time
    heatmap_data = np.zeros((num_qubits, t_steps))

    # V is always X on the first qubit (The "Butterfly")
    op_list_V = [np.eye(2)] * num_qubits
    op_list_V[0] = np.array([[0, 1], [1, 0]]) # Pauli X
    V = op_list_V[0]
    for i in range(1, num_qubits):
        V = np.kron(V, op_list_V[i])

    #print(f"Generating Heatmap (h_z={h_z})...")

    for t_idx, t in enumerate(times):
        if t == 0: continue

        # Get the global Unitary for this time point
        U = get_unitary_from_circuit(qubits, t, h_x, h_z) # Using previous function
        U_dag = U.conj().T

        # Calculate OTOC for each qubit j as the "Probe"
        for j in range(num_qubits):
            # W is Z on qubit j
            op_list_W = [np.eye(2)] * num_qubits
            op_list_W[j] = np.array([[1, 0], [0, -1]]) # Pauli Z

            # Construct W matrix
            W = op_list_W[0]
            for k in range(1, num_qubits):
                W = np.kron(W, op_list_W[k])

            # Heisenberg evolution: W(t) = U^dag * W * U
            W_t = U_dag @ W @ U

            # Commutator magnitude
            commutator = W_t @ V - V @ W_t
            heatmap_data[j, t_idx] = np.linalg.norm(commutator) ** 2

    return times, heatmap_data

# --- Execution ---
n_qubits = 6 # Kept small for simulation speed
t_max = 20.0
t_steps = 25

# Generate data for Chaotic vs Ordered
_, data_ordered = calculate_otoc_heatmap(n_qubits, 1.0, 0.0, t_max, t_steps)
times, data_chaotic = calculate_otoc_heatmap(n_qubits, 1.0, 0.5, t_max, t_steps)

# --- Plotting ---

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 3))

# Define your desired font size
small_size = 7

# Plot 1
sns.heatmap(data_ordered, ax=ax1, cmap='magma', xticklabels=np.round(times, 1))
ax1.set_title("Ordered (h_z=0): Ballistic Spread", fontsize=small_size + 2)
ax1.set_xlabel("Time (t)", fontsize=small_size)
ax1.set_ylabel("Qubit Index", fontsize=small_size)
ax1.tick_params(axis='both', labelsize=small_size)

# Adjust Colorbar 1
cbar1 = ax1.collections[0].colorbar
cbar1.ax.tick_params(labelsize=small_size)

# Plot 2
sns.heatmap(data_chaotic, ax=ax2, cmap='magma', xticklabels=np.round(times, 1))
ax2.set_title("Chaotic (h_z=0.5): Scrambling / Diffusion", fontsize=small_size + 2)
ax2.set_xlabel("Time (t)", fontsize=small_size)
ax2.set_ylabel("Qubit Index", fontsize=small_size)
ax2.tick_params(axis='both', labelsize=small_size)

# Adjust Colorbar 2
cbar2 = ax2.collections[0].colorbar
cbar2.ax.tick_params(labelsize=small_size)

plt.tight_layout()
plt.show()

display(Markdown(r"*$C(t) = \langle [\hat{W}(t), \hat{V}(0)]^\dagger [\hat{W}(t), \hat{V}(0)] \rangle$ quantifies rate of scrambling. $\hat{V}(0)$: Butterfly (small local perturbation at start). $\hat{W}(t)$ measurement later in time. At $t=0$, $W$ and $V$ commute (are on different qubits). Commutator is 0. As time evolves, $\hat{W}(t)$ grows spatially (scrambling). Eventually, $\hat{W}(t)$ hits the qubit where $\hat{V}$ sits. They no longer commute. Commutator grows, and $C(t)$ deviates from zero. Result: rate at which OTOC grows is Lyapunov Exponent ($\lambda$), the speed limit of chaos.*"))
```

<font color="blue">*Variationsanalyse in Quantum Mechanics*

* **Energetische Variationsprinzip**: jeder (normierte) Zustand $\psi$ hat Energieerwartung, die größer oder gleich der tatsächlichen Grundzustandsenergie $E_0$ ist.
  * Energie als Funktional $E[\psi]$ (Funktion auf Zahl): $E[\psi] = \frac{\langle \psi | \hat{H} | \psi \rangle}{\langle \psi | \psi \rangle} \ge E_0$.
  * Natur versucht, Zustand niedrigster Energie einzunehmen. $\psi$ variieren ($\delta E = 0$) um stationäre Zustände zu finden.Suche nach Minimum findet im Sobolev-Raum $H^1$ statt, da hier der Erwartungswert der kinetischen Energie wohldefiniert ist.
* **Rayleigh-Ritz-Verfahren**: Schrödinger-Gleichung für komplexe Systeme (Atome mit vielen Elektronen) nicht exakt lösbar.
  * Variationsanalyse als Näherungen. Man wählt eine Familie von Testfunktionen $\psi(\alpha, \beta, \dots)$ mit variablen Parametern. Man berechnet $E(\alpha, \beta, \dots)$ und minimiert die Parameter: $\frac{\partial E}{\partial \alpha} = 0, \quad \frac{\partial E}{\partial \beta} = 0, \dots$. Man erhält obere Schranke für wahre Energie HF, DFT).
* **Schrödinger-Gleichung ist die Euler-Lagrange-Gleichung** eines spezifischen Variationsproblems.
  * Wenn man das Funktional $J[\psi] = \langle \psi | \hat{H} | \psi \rangle$ unter der Nebenbedingung der Normierung $\langle \psi | \psi \rangle = 1$ minimiert (mittels Lagrange-Multiplikatoren), folgt direkt: $\hat{H}\psi = E\psi$.
  * Eigenwertprobleme der Funktionalanalysis = Minimierungsprobleme der Variationsrechnung.

<font color="blue">*Funktionentheorie (Complex Analysis) in Quantum Mechanics*

* **Holomorphie (komplex differenzierbar)** ist Bedingung für unendliche Glätte. In QM ist das Wächter über physikalische Sinnhaftigkeit.
  * Cauchy-Riemann-Gleichungen: Für $f(x+iy) = u + iv$ muss gelten: $\frac{\partial u}{\partial x} = \frac{\partial v}{\partial y}, \quad \frac{\partial u}{\partial y} = -\frac{\partial v}{\partial x}$.
  * **Ganze Funktionen & Physikalische Energie:** viele Lösungen (z. B. der Harmonische Oszillator) sind ganze Funktionen (überall auf $\mathbb{C}$ holomorph). Das hat fundamentale Konsequenzen: **Ausschluss von „Knicken“:** Da der Hamilton-Operator $\hat{H} \propto \frac{d^2}{dx^2}$ die zweite Ableitung enthält, muss $\psi$ glatt sein. Ein „Knick“ (Diskontinuität der Ableitung) würde zu einer unendlichen kinetischen Energie führen.
* **Beispiel Harmonischer Oszillator:** Die Lösung $\psi_n(x) = H_n(x) e^{-x^2/2}$ besteht aus Hermite-Polynomen und der Gauß-Glocke – beides ganze Funktionen.
  * Ihre Holomorphie garantiert, dass die Wellenfunktion weich abklingt und keine physikalisch unsinnigen Sprünge macht.
  * **Identitätssatz & Starrheit:** Wenn eine holomorphe Wellenfunktion auf einem winzigen Bereich bekannt ist, ist sie durch ihre analytische Fortsetzung im gesamten Raum festgelegt. Dies verhindert „instantane Teleportation“ oder lokale Änderungen ohne globale Auswirkungen.
  * **Erhaltungsgrößen:** Ohne Differenzierbarkeit wären der Impulsoperator $\hat{p} = -i\hbar \nabla$ und der Wahrscheinlichkeitsstrom $j$ (der die erste Ableitung nutzt) an Singulärstellen nicht definiert.
* **Wick-Rotation**: Die Funktionentheorie erlaubt es, die Quantendynamik mit der statistischen Mechanik zu verknüpfen, indem man die Zeitkoordinate komplex auffasst.
  * **Imaginäre Zeit:** Durch die Ersetzung $t = -i\tau$ transformiert sich die Schrödinger-Gleichung in eine Diffusions- oder Wärmeleitungsgleichung: $e^{-i\hat{H}t/\hbar} \quad \xrightarrow{t \to -i\tau} \quad e^{-\hat{H}\tau/\hbar}$.
  * Der Zeitentwicklungsoperator wird formal identisch mit dem Boltzmann-Faktor der Thermodynamik. Dies ist die Basis für **euklidische Pfadintegrale**, die in der Gitter-QCD und bei der Berechnung von Hawking-Strahlung entscheidend sind.


<font color="blue">*Funktionalanalysis (Functional Analysis) in Quantum Mechanics*

* Klassische Mechanik mit Punkten im Phasenraum. Quantenmechanik in unendlichdimensionalen Funktionenräumen: Zustandsraum: $L^2(\mathbb{R}^n)$ und Hilbert-Räume. Zustand eines Systems durch Einheitsvektor $\psi$ in komplexen Hilbert-Raum $\mathcal{H}$ dargestellt.
  * **Quadratintegrabilität:** Um sicherzustellen, dass die Gesamtwahrscheinlichkeit, ein Teilchen zu finden, genau 1 ergibt, muss $\psi \in L^2(\mathbb{R}^n)$ gelten: $\int_{\mathbb{R}^n} |\psi(x)|^2 \, dx < \infty$
  * **Skalarprodukt:** Die Born’sche Regel basiert auf dem Hilbert-Skalarprodukt $\langle \phi | \psi \rangle = \int \phi^* \psi \, dx$. Dies ermöglicht die Berechnung von Wahrscheinlichkeitsamplituden und Projektionen auf bestimmte Zustände.
  * **Vollständigkeit:** Ein Hilbert-Raum muss vollständig sein (jede Cauchy-Folge von Funktionen konvergiert gegen eine Funktion innerhalb des Raums). Dies garantiert, dass Grenzwerte physikalischer Prozesse weiterhin physikalisch valide Zustände darstellen.
* **Observablen als selbstadjungierte Operatoren**: Jede messbare physikalische Größe (Observable) entspricht einem linearen Operator, der auf $\mathcal{H}$ wirkt.
  * **Selbstadjungiertheit**: In endlichen Dimensionen reicht „hermitesch“ aus. In unendlichen Dimensionen müssen Operatoren selbstadjungiert sein ($A = A^\dagger$ unter Berücksichtigung des Definitionsbereichs $D(A)$). Dies ist entscheidend, da der Spektralsatz garantiert, dass selbstadjungierte Operatoren reelle Eigenwerte haben – die einzig möglichen Ergebnisse einer Messung.
  * **Zentrale Operatoren:** Impuls: $\hat{p} = -i\hbar \nabla$. Hamilton-Operator (Energie): $\hat{H} = -\frac{\hbar^2}{2m}\Delta + V(x)$ **Vertauschungsrelationen:** Die Tatsache, dass $x$ und $p$ nicht kommutieren ($[\hat{x}, \hat{p}] = i\hbar$), führt direkt zur Heisenbergschen Unschärferelation – eine fundamentale Konsequenz der Operatoreigenschaften in Hilbert-Räumen.
* **Sobolev-Räume und schwache Ableitungen**: Physikalische Potentiale sind oft nicht glatt (z. B. das Delta-Potential $\delta(x)$ oder der unendliche Potentialtopf).
  * **Schwache Ableitungen:** Klassische Ableitungen versagen an „Knicken“ der Wellenfunktion. Die Funktionalanalysis führt Sobolev-Räume $H^k(\mathbb{R}^n)$ ein, in denen Funktionen Ableitungen im distributionellen Sinne besitzen.
  * **Wohldefiniertheit der Energie:** Damit die kinetische Energie $\langle \psi | -\Delta | \psi \rangle$ endlich bleibt, muss $\psi$ im Sobolev-Raum $H^1(\mathbb{R}^n)$ liegen. Dies stellt sicher, dass selbst „spitze“ Wellenfunktionen einen mathematisch präzisen Energiewert haben.
* **Unitäre Entwicklung und Fourier-Transformation**: Dynamik eines Quantensystems wird durch **unitäre Operatoren** beschrieben, welche die Norm (und damit die Wahrscheinlichkeit) erhalten.
  * Nach dem Satz von Stone ist der **Zeitentwicklungsoperator** $U(t) = e^{-i\hat{H}t/\hbar}$ genau dann unitär, wenn $\hat{H}$ selbstadjungiert ist. Dies „rotiert“ den Zustandsvektor im Hilbert-Raum, ohne seine Länge zu verändern.
  * **Fourier-Transformation als unitäre Abbildung:** Der Wechsel zwischen Ortsraum ($x$) und Impulsraum ($p$) ist eine unitäre Transformation (Satz von Plancherel): $\tilde{\psi}(p) = \mathcal{F}\{\psi(x)\}$ =beides physikalisch äquivalente Beschreibungen desselben Vektors im Hilbert-Raum.




<font color="blue">**Mathematical Formulation of Quantum Mechanics**

*Quantum Complexity*

* **$\mathcal{O}(n^2)$ and $\mathcal{O}(\log^2 d)$ are the exact same thing.**
* Let $n$ be the number of qubits.
* The dimension of the Hilbert space $d$ grows exponentially with the number of qubits: $d = 2^n$.
* Therefore, taking the logarithm gives you the number of qubits: $\log_2(d) = n$.
* So, squaring it: $\mathcal{O}(\log^2 d) = \mathcal{O}(n^2)$.

0 \\
1
\end{array}\right)$. This state cannot be factored as $|a\rangle \otimes|b\rangle$. So this is genuine entanglement.
  * Case 2:  One Qubit in Superposition, One in a Basis State. $\left(\frac{|0\rangle+|1\rangle}{\sqrt{2}}\right) \otimes|0\rangle=\frac{|00\rangle+|10\rangle}{\sqrt{2}}$.  Vector form: $
\binom{\frac{1}{\sqrt{2}}}{\frac{1}{\sqrt{2}}} \otimes\binom{1}{0}=\left(\begin{array}{c}
\frac{1}{\sqrt{2}} \\
0 \\
\frac{1}{\sqrt{2}} \\
0
\end{array}\right)$. This is still a product state (not entangled).
5. **Two Qubits: Bell state** $|\Phi^+\rangle$ (Entangled State / Superdense Coding)
  * First, Hadamard on $q_0$: $|00\rangle \xrightarrow{H \otimes I} \left(\frac{|0\rangle + |1\rangle}{\sqrt{2}}\right) \otimes |0\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |10\rangle)$
  * Second, CNOT on $q_1$: $\frac{1}{\sqrt{2}}(|00\rangle + |10\rangle) \xrightarrow{\text{CNOT}} \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle) = |\Phi^+\rangle$
  * The amplitude for $|00\rangle$ is $1/\sqrt{2}$. The probability $P('00') = |1/\sqrt{2}|^2 = 0.5$
  * The amplitude for $|11\rangle$ is $1/\sqrt{2}$. The probability $P('11') = |1/\sqrt{2}|^2 = 0.5$
  * Bell state $|\Phi^+\rangle$ = $\frac{1}{\sqrt{2}}|00\rangle+\frac{1}{\sqrt{2}}|11\rangle$ can not be written as a product of two single-qubit states
6. **Multi-Qubit Amplitude Expansion** (Distributive Property of Superposition)
  * The total state of a multi-qubit system is the tensor product of individual qubit operations.
  * Expanding a superposition state means distributing the tensor product across all terms:
  * Initial State: $|\psi_1\rangle = |0\rangle_{q_3} \otimes |0\rangle_{q_2} \otimes \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)_{q_1} \otimes (\frac{\sqrt{3}}{2}|0\rangle + \frac{1}{2}|1\rangle)_{q_0}$
  * Expansion: $|\psi_1\rangle = \frac{\sqrt{3}}{2\sqrt{2}}|0000\rangle + \frac{1}{2\sqrt{2}}|0001\rangle + \frac{\sqrt{3}}{2\sqrt{2}}|0010\rangle + \frac{1}{2\sqrt{2}}|0011\rangle$
  * Basis State Notation: In $|q_3 q_2 q_1 q_0\rangle$, each digit corresponds to the state of a specific qubit (e.g., in $|0010\rangle$, $q_1$ is in state $|1\rangle$).
7. **Multi-Qubit Tensor Products**
  * Tensor Product: with multiple qubits, we describe the total system using the **tensor product** (denoted by $\otimes$).
  * Notation $|\psi_0\rangle = |0000\rangle$ is a shorthand for: $|\psi_0\rangle = |0\rangle_{q_3} \otimes |0\rangle_{q_2} \otimes |0\rangle_{q_1} \otimes |0\rangle_{q_0}$.
  * Means "qubit 3 is in the state $|0\rangle$, AND qubit 2 is in the state $|0\rangle$, and..." and so on. Compact label $|0000\rangle$ is better to read.
  * Writing this out with the full vector math, we would be taking the tensor product of four 2-dimensional vectors.
  * So: $|0\rangle$ is 2-dimensional vector state $\begin{pmatrix} 1 \\ 0 \end{pmatrix}$. $|0000\rangle$ is 16-dimensional vector for four qubits all in $|0\rangle$ state.

<font color="blue">**Quantum Measurements**

1. Math Basics: Bra-Ket $\langle \phi | \psi \rangle$ vs Ket-Bra $P = |\phi\rangle\langle\phi|$**

* **Overlap/Amplitude** (Bra-Ket) $\langle \phi | \psi \rangle$: is a **Number** (Amplitude). Used for Basic building block for probabilities.
* **Measurement Operator** (Ket-Bra) $P = |\phi\rangle\langle\phi|$ is an **Operator** (Matrix) used to define measurement filters (projectors) + quantum states (density matrix) A "Filter" that checks for state $|\phi\rangle$. (Rank-1 Projector). Represents standard measuring device.

2. Measurements on Pure States: Vectors $|\psi\rangle$ with inner product $P_i = |\langle \phi_i | \psi \rangle|^2$ or sandwich $P(i) = \langle \psi | \hat{P}_i | \psi \rangle$**

> If operators $\{M_i\}$ are orthogonal projectors $\{P_i\}$, it describes **PVM** for ideal, sharp measurements. $P_i^2 = P_i$, $P_i P_j = \delta_{ij} P_i$. Pure States $|\psi\rangle$ can use "Sandwich" $\langle\psi| P |\psi\rangle$ to calculate probability. Observable $A$ defines these projectors ($A = \sum a_i P_i$).

* **Probability (Standard / Simple)** Probability of outcome $i$ for state  $P_i = |\langle \phi_i | \psi \rangle|^2$: (**Born Rule**). Simple square of amplitude.
  * Probability of finding specific state $|\psi\rangle$ in state $|\phi\rangle$. Valid when projecting onto basis vectors.
  * Formally, there is an operator sandwiched in between: $\langle \psi | \hat{P} | \psi \rangle$, but this is in this case only the identity operator $\hat{P}$ which we can ignore: $P(i) = \langle \psi | \hat{I}_i | \psi \rangle = \langle \psi | \psi \rangle$
  * Even in this simple case, we are technically using the sandwich method! The operator $\hat{P}$ is the projector $|\phi\rangle\langle\phi|$. We can associate the bra and ket to scalar products: $\langle \psi | \phi \rangle \cdot \langle \phi | \psi \rangle$. Result: $|\langle \phi | \psi \rangle|^2$.
  * If the operator were the Identity $\hat{I}$, the result would be $\langle \psi | I | \psi \rangle = \langle \psi | \psi \rangle = 1$. This means the probability is 100%. This only happens if you are asking "Does the particle exist?" or if you sum up all possible outcomes ($\sum P_i = I$).
* **Probability (General / Sandwich)** Probability of outcome $i$ for state  $P(i) = \langle \psi | \hat{P}_i | \psi \rangle$: for any kind of projective measurement:
  * **Subspaces (Multi-dimensional)**: $\rightarrow$ **PVM** with $P_i^2 = P_i$ (Degenerate/Coarse-grained). Projects onto Subspace (Multi-D plane). Sometimes an outcome corresponds to whole group of states ("Energy > 0"), not just single state vector, we bundle outcomes together ("Any state in x-y plane" or "Is energy level 2? Could be state $2a, 2b$, or $2c$"). Is coarse-graining, but it is still sharp/projective PVM. Can't use simple amplitude square $|\langle \phi | \psi \rangle|^2$ - there is no single $|\phi\rangle$. Must use sandwich $\langle \psi | \hat{P}_{group} | \psi \rangle$.
  * **Single Vectors (One-dimensional)**: $\rightarrow$ **PVM** with $P_i^2 = P_i$ (Rank-1/Fine-grained). Rank-1 PVM: Projects onto a Single Vector (1D line). "Is the spin exactly Up?" If you are measuring a single state $|\phi\rangle$ (=measurement is simple rank-1 projector onto specific state $|\phi\rangle$), the operator is $\hat{P} = |\phi\rangle\langle\phi|$ and the general formula naturally simplifies back to the specific one: $\langle \psi | (|\phi\rangle\langle\phi|) | \psi \rangle = \langle \psi | \phi \rangle \langle \phi | \psi \rangle = |\langle \phi | \psi \rangle|^2$
* **Expectation Value**: $\langle A \rangle = \langle \psi | \hat{A} | \psi \rangle$. The average outcome over many runs. Note that the "sandwich" is a trace (like in Mixed States)! For a pure state where $\rho = |\psi\rangle\langle\psi|$: $\text{Tr}(\rho P) = \text{Tr}(|\psi\rangle\langle\psi| P) = \langle\psi| P |\psi\rangle$ due to cyclic property of trace: $\text{Tr}(AB) = \text{Tr}(BA)$

3. **Measurements on Mixed States: Density Matrix $\rho = \sum p_i |\psi_i\rangle\langle\psi_i|$ with Trace $P_i = \text{Tr}(\rho \hat{E}_i)$**

>If operators $\{M_i\}$ are positive operators, it describes **POVM** for noisy measurements. $M_i \geq 0$, $\sum M_i = I$. Mixed States must use Trace with POVM $P_i = \text{Tr}(\rho P)$ because there is no single vector $|\psi\rangle$ but probability distribution of vectors, represented by matrix $\rho = \sum p_i |\psi_i\rangle\langle\psi_i|$. We cannot "sandwich" a matrix inside a vector if we don't *have* a single vector.

* **Probability (Trace Rule)** Probability of outcome $i$ for state $P_i = \text{Tr}(\rho \hat{M}_i)$: The universal formula. "Ket-Bra" structure is the more fundamental, universal way to describe both states and measurements in standard quantum mechanics.
  * **PVM (Projective)**: If $\hat{M}_i$ are orthogonal projectors ($P_i P_j = \delta_{ij} P_i$). Ideal measurement.
  * **POVM (Generalized)**: If $\hat{M}_i$ positive operators summing to Identity ($\sum E_i = I$). Noisy or ambiguous. Relaxes orthogonality.
* **Expectation Value:** $\langle A \rangle = \text{Tr}(\rho \hat{A})$.
* **Notation**: POVM "Element" vs. "Kraus Operator" we used $E_i$ or $M_i$. In quantum algorithms, we distinguish between:
  * POVM Element ($E_i$) to **calculate probabilities**: $E_i = M_i^\dagger M_i$ for $P(i) = \text{Tr}(\rho E_i)$
  * Measurement Operator ($M_i$, or Kraus Operator) to **calculate state change** (post-measurement state): $\rho_{new} = \frac{M_i \rho M_i^\dagger}{\text{Tr}(\rho E_i)}$
  * We focus on probability ($E_i$). But if we need to know "what happens to qubit after measurement," we need $M_i$, not just $E_i$.

4. **Special: Sandwich Usage Restriction only for States (only Pure), but not. for Operators (can be anything)**

* **Special: Non-Orthogonal / Overcomplete States** $\rightarrow$ **POVM** (Noisy/Ambiguous) but can use sandwich $\langle \psi | \hat{P}_i | \psi \rangle$.
  * You can also sandwich POVM. "sandwich is only for pure states" - but that **is true regarding the State, not the Operator**.
  * Restriction: You cannot sandwich a Density Matrix ($\rho$) inside a vector. But you can sandwich any square operator (Projector $P$, POVM element $E$, Hamiltonian $H$) inside a vector $|\psi\rangle$. So, if you are working with a simple vector $|\psi\rangle$ and a messy POVM element $\hat{E}_i$, you are allowed to just sandwich it: $P(i) = \langle \psi | \hat{E}_i | \psi \rangle$.
  * For Pure States, we can calculate the probability of a POVM outcome using the sandwich: $P(i) = \langle \psi | \hat{E}_i | \psi \rangle$. This is valid even though $\hat{E}_i$ is not a sharp projector ($E_i^2 \neq E_i$), as long as it is a positive operator."
  * This is not a projective measurement, but "Generalized" measurement. $P(i) = \langle \psi | \hat{P}_i | \psi \rangle$ is still valid when $P_i$ is just a Positive Operator ($E_i$), not a Projector ($P_i^2 = P_i$). $\sum E_i = I$.
  * Used when outcomes are Not Orthogonal and we have ambiguity where outcomes not sharp, projective but overlap ("trying to distinguish 3 states in a 2D circle" or "Mostly x-direction" vs "Mostly 45-degree direction"). Using Trace is cleaner though.

5. **Advanced Measurements**

* **Weak Value** $\langle A \rangle_w = \frac{\langle \phi_f | A |\psi_i \rangle}{\langle \phi_f | \psi_i \rangle}$: The "conditional average" between a pre-selected state $|\psi_i\rangle$ and a post-selected state $|\phi_f\rangle$. Complex-valued! Used for conditional properties without full state collapse.

*<font color="blue">Appendix Check: Heisenberg Picture*

* Schrödinger Picture: State rotates $|\psi\rangle \to U|\psi\rangle$.
* Heisenberg Picture: Ruler rotates $A \to U^\dagger A U$.
* Result: The expectation value $\langle \psi | U^\dagger A U | \psi \rangle$ is identical in both views.
